## Trimming the data

64 columns feels unwieldy, so I'm going to cut it down and make it a new data frame called **walkbike_trimmed**.

````python
walkbike_trimmed = walkbike[['Facility_Type',
                            'Ped_Total',
                            'Ped_Female',
                            'Ped_MobAssist',
                            'Ped_Stroller',
                            'Bike_Total',
                            'Bike_Female',
                            'Bike_Sidewalk',
                            'Bike_ContraFlow',
                            'Bike_Bikeshare',
                            'Bike_Helmet',
                            'Other_Total',
                            'Other_Female',
                            'Other_Sidewalk',
                            'Other_ContraFlow',
                            'Other_Helmet',
                            'Scooter_Total',
                            'Scooter_Female',
                            'Scooter_Sidewalk',
                            'Scooter_ContraFlow',
                            'Scooter_Helmet']].copy()
walkbike_trimmed
````

Next, I want to convert the ```Facility_Type``` column into a string and all the rest of the columns into integers so I can make plots. My intitial attempt of using the ```convert_dtypes()``` function I found online didn't quite work the way I hoped.

````python
walkbike_trimmed = walkbike_trimmed.convert_dtypes()

walkbike_trimmed.dtypes
````

Well, that made everything into strings, except for ```Ped_Total``` which I had already converted to an integer earlier. Let's try again, with help from [Geeks for Geeks](https://www.geeksforgeeks.org/python-pandas-dataframe-loc/) and using a for loop:

````python
for col in walkbike_trimmed.loc[:, 'Ped_Female':'Scooter_Helmet']:
    walkbike_trimmed[col] = pd.to_numeric(walkbike[col])
````

Looks like this made all the columns from ```Ped_Female``` onward floats, which I think works for now:

````python
walkbike_trimmed.info()
````
